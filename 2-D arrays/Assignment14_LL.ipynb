A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

</aside>
Input:
LinkedList: 4->5->6
Output:457
Example 2:
    Input:
LinkedList: 1->2->3
Output:124

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverseLinkedList(head):
    prev = None
    curr = head

    while curr:
        nextNode = curr.next
        curr.next = prev
        prev = curr
        curr = nextNode

    return prev

def addOne(head):
    # Reverse the linked list
    reversedHead = reverseLinkedList(head)

    curr = reversedHead
    carry = 1

    while curr:
        # Add the carry to the current node
        curr.data += carry

        # Update the carry
        carry = curr.data // 10
        curr.data = curr.data % 10

        # Move to the next node
        prev = curr
        curr = curr.next

    # If there's still a carry, create a new node with a value of 1
    if carry:
        newNode = Node(1)
        prev.next = newNode

    # Reverse the modified linked list
    modifiedHead = reverseLinkedList(reversedHead)

    return modifiedHead

# Function to create a linked list from an array of values
def createLinkedList(values):
    if not values:
        return None

    head = Node(values[0])
    curr = head

    for val in values[1:]:
        newNode = Node(val)
        curr.next = newNode
        curr = curr.next

    return head

# Function to convert the linked list to an integer
def linkedListToInt(head):
    curr = head
    num = 0

    while curr:
        num = num * 10 + curr.data
        curr = curr.next

    return num

# Example 1
values1 = [4, 5, 6]
head1 = createLinkedList(values1)
print("Linked List: ", linkedListToInt(head1))
newHead1 = addOne(head1)
print("Modified Linked List: ", linkedListToInt(newHead1))


# Example 2
values2 = [1, 2, 3]

head2 = createLinkedList(values2)
print("Linked List: ", linkedListToInt(head2))
newHead2 = addOne(head2)
print("Modified Linked List: ", linkedListToInt(newHead2))
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def oddEvenList(head):
    if not head or not head.next:
        return head

    odd_head = head
    even_head = head.next
    odd = odd_head
    even = even_head

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head

    return odd_head


# Example 1
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(3)
head1.next.next.next = ListNode(4)
head1.next.next.next.next = ListNode(5)

print("Example 1:")
print("Input:", end=" ")
curr = head1
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()

reordered_head1 = oddEvenList(head1)

print("Output:", end=" ")
curr = reordered_head1
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()
print()

# Example 2
head2 = ListNode(2)
head2.next = ListNode(1)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(5)
head2.next.next.next.next = ListNode(6)
head2.next.next.next.next.next = ListNode(4)
head2.next.next.next.next.next.next = ListNode(7)

print("Example 2:")
print("Input:", end=" ")
curr = head2
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()

reordered_head2 = oddEvenList(head2)

print("Output:", end=" ")
curr = reordered_head2
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()
````

**Example 1:**

```"

Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2
Rotate 2:Â 7Â -> 8Â -> 9Â -> 2Â -> 4

```

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def leftShift(head, k):
    if not head or not head.next or k == 0:
        return head

    # Find the length of the linked list
    length = 0
    curr = head
    while curr:
        length += 1
        curr = curr.next

    # Calculate the actual number of shifts needed
    k = k % length

    if k == 0:
        return head

    # Find the (k+1)-th node from the end
    fast = head
    for _ in range(k):
        fast = fast.next

    slow = head
    while fast.next:
        slow = slow.next
        fast = fast.next

    # Perform the left shift
    new_head = slow.next
    slow.next = None
    fast.next = head

    return new_head


# Example 1
head1 = ListNode(2)
head1.next = ListNode(4)
head1.next.next = ListNode(7)
head1.next.next.next = ListNode(8)
head1.next.next.next.next = ListNode(9)
k1 = 3

print("Example 1:")
print("Input:", end=" ")
curr = head1
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()

shifted_head1 = leftShift(head1, k1)

print("Output:", end=" ")
curr = shifted_head1
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()
print()

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(5)
head2.next.next.next.next.next = ListNode(6)
head2.next.next.next.next.next.next = ListNode(7)
head2.next.next.next.next.next.next.next = ListNode(8)
k2 = 4

print("Example 2:")
print("Input:", end=" ")
curr = head2
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()

shifted_head2 = leftShift(head2, k2)

print("Output:", end=" ")
curr = shifted_head2
while curr:
    print(curr.val, end=" ")
    curr = curr.next
print()

````

For each node in the list, find the value of theÂ **next greater node**. That is, for each node, find the value of the first node that is next to it and has aÂ **strictly larger**Â value than it.

Return an integer arrayÂ `answer`Â whereÂ `answer[i]`Â is the value of the next greater node of theÂ `ith`Â node (**1-indexed**). If theÂ `ith`Â node does not have a next greater node, setÂ `answer[i] = 0`.

**Example 1:**

```
Input: head = [2,1,5]
Output: [5,5,0]

```

**Example 2:**

!https://assets.leetcode.com/uploads/2021/08/05/linkedlistnext2.jpg

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def nextLargerNodes(head):
    # Convert the linked list to an array
    values = []
    curr = head
    while curr:
        values.append(curr.val)
        curr = curr.next

    n = len(values)
    result = [0] * n
    stack = []

    for i in range(n):
        while stack and values[i] > values[stack[-1]]:
            result[stack.pop()] = values[i]
        stack.append(i)

    return result


# Example 1
head1 = ListNode(2)
head1.next = ListNode(1)
head1.next.next = ListNode(5)

print("Example 1:")
print("Input: [2, 1, 5]")
print("Output:", nextLargerNodes(head1))
print()

# Example 2
head2 = ListNode(2)
head2.next = ListNode(7)
head2.next.next = ListNode(4)
head2.next.next.next = ListNode(3)
head2.next.next.next.next = ListNode(5)

print("Example 2:")
print("Input: [2, 7, 4, 3, 5]")
print("Output:", nextLargerNodes(head2))
````

Given theÂ `head`Â of a linked list, we repeatedly delete consecutive sequences of nodes that sum toÂ `0`Â until there are no such sequences.

After doing so, return the head of the final linked list.Â  You may return any such answer.

(Note that in the examples below, all sequences are serializations ofÂ `ListNode`Â objects.)

**Example 1:**

```
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```

**Example 2:**

```
Input: head = [1,2,3,-3,4]
Output: [1,2,4]

```

**Example 3:**

```
Input: head = [1,2,3,-3,-2]
Output: [1]
```

</aside>

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head
    stack = []
    prefix_sum = 0
    curr = dummy

    while curr:
        prefix_sum += curr.val

        if prefix_sum in stack:
            # Remove nodes between curr and the node with the same prefix_sum
            to_remove = stack[stack.index(prefix_sum) + 1:]
            for node in to_remove:
                node.next = None
            stack = stack[:stack.index(prefix_sum) + 1]

        else:
            stack.append(curr)

        curr = curr.next

    return dummy.next
````

````python
# Example 1
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(-3)
head1.next.next.next = ListNode(3)
head1.next.next.next.next = ListNode(1)

result1 = removeZeroSumSublists(head1)
output1 = []
while result1:
    output1.append(result1.val)
    result1 = result1.next

print("Example 1:")
print("Input: [1, 2, -3, 3, 1]")
print("Output:", output1)
print()

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(-3)
head2.next.next.next.next = ListNode(4)

result2 = removeZeroSumSublists(head2)
output2 = []
while result2:
    output2.append(result2.val)
    result2 = result2.next

print("Example 2:")
print("Input: [1, 2, 3, -3, 4]")
print("Output:", output2)
print()

# Example 3
head3 = ListNode(1)
head3.next = ListNode(2)
head3.next.next = ListNode(3)
head3.next.next.next = ListNode(-3)
head3.next.next.next.next = ListNode(-2)

result3 = removeZeroSumSublists(head3)
output3 = []
while result3:
    output3.append(result3.val)
    result3 = result3.next

print("Example 3:")
print("Input: [1, 2, 3, -3, -2]")
print("Output:", output3)
print()
````
