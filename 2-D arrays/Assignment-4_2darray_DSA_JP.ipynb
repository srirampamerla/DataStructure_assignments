**Question 1**
Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** the integers that appeared in **all** three arrays.

**Example 1:**

Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

Output: [1,5]

**Explanation:** Only 1 and 5 appeared in the three arrays.


```python
def find_common_elements(arr1, arr2, arr3):
    # Initialize three pointers, one for each array
    ptr1 = ptr2 = ptr3 = 0
    
    common_elements = []
    
    # Iterate until one of the pointers reaches the end of its respective array
    while ptr1 < len(arr1) and ptr2 < len(arr2) and ptr3 < len(arr3):
        # If the current elements at all three pointers are equal, it is a common element
        if arr1[ptr1] == arr2[ptr2] == arr3[ptr3]:
            common_elements.append(arr1[ptr1])
            # Move all three pointers forward
            ptr1 += 1
            ptr2 += 1
            ptr3 += 1
        # If the current element in arr1 is smaller, move ptr1 forward
        elif arr1[ptr1] < arr2[ptr2] or arr1[ptr1] < arr3[ptr3]:
            ptr1 += 1
        # If the current element in arr2 is smaller, move ptr2 forward
        elif arr2[ptr2] < arr1[ptr1] or arr2[ptr2] < arr3[ptr3]:
            ptr2 += 1
        # If the current element in arr3 is smaller, move ptr3 forward
        else:
            ptr3 += 1
    
    return common_elements
```

```python
#n is the total number of elements in the three arrays. We iterate through the arrays using the three pointers
print("The time complexity is O(n)")
print("Space COmplexity is O(1)")
arr1 = [1,2,3,4,5]
arr2 = [1,2,5,7,9]
arr3 = [1,3,4,5,8]
find_common_elements(arr1, arr2, arr3)
```

**Question 2**

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].


```python
# Approach -1
def find_disjoint_nums(nums1, nums2):
    distinct_nums1 = []  # List to store distinct integers in nums1 not present in nums2
    distinct_nums2 = []  # List to store distinct integers in nums2 not present in nums1
    
    # Iterate over nums1 and check if each integer is present in nums2
    for num in nums1:
        if num not in nums2:
            distinct_nums1.append(num)  # Add the distinct integer to distinct_nums1
    
    # Iterate over nums2 and check if each integer is present in nums1
    for num in nums2:
        if num not in nums1:
            distinct_nums2.append(num)  # Add the distinct integer to distinct_nums2
    
    return [distinct_nums1, distinct_nums2]

print("The time complexity is O(m*n)")
print("Space COmplexity is O(m+n)")
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
find_disjoint_nums(nums1, nums2)
```

**Question 3**
Given a 2D integer array matrix, return *the **transpose** of* matrix.

The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

**Example 1:**

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [[1,4,7],[2,5,8],[3,6,9]]


```python
def transpose(matrix):
    # Get the number of rows and columns in the matrix
    rows = len(matrix)
    cols = len(matrix[0])
    transposed_matrix=[]
    
    # Create a new matrix with swapped dimensions (cols x rows)
    # transposed_matrix = [[0 for i in range(rows)] for j in range(cols)]
    # Iterate over the columns
    for j in range(cols):
        column = []
    
    # Iterate over the rows
        for i in range(rows):
            column.append(0)
    
        transposed_matrix.append(column)
    
    # Iterate over the elements of the original matrix and assign them to the transposed matrix
    for i in range(rows):
        for j in range(cols):
            transposed_matrix[j][i] = matrix[i][j]
    
    return transposed_matrix
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
transpose(matrix)
```

**Question 4**
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is **maximized**. Return *the maximized sum*.

**Example 1:**

Input: nums = [1,4,3,2]

Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3

2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3

3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

So the maximum possible sum is 4.

```python
def Sum_2(nums):
    nums.sort()  # Sorting the input list in ascending order
    if(len(nums))==0:
        return 0  # If the list is empty, return 0
    result = 0
    for i in range(0, len(nums)-1, 2):
        result += min(nums[i], nums[i+1])  # Adding the minimum value between current number and next number to the result

    return result
nums = [1,4,3,2]
Sum_2(nums)
```

**Question 5**
You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**

[]()

**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.

```python
def arrangeCoins(n):
    rows = 0  # Initialize the variable to count the number of complete rows
    while n >= rows + 1:  # Continue until the remaining coins are enough to form the next row
        rows += 1 # Increment the row count
        n -= rows# Subtract the coins used to form the current row from the remaining coins
    return rows  # Return the number of complete rows formed
n=5
arrangeCoins(n)
```

```python
print("The time complexity is O(n)")
print("Space COmplexity is O(1)")
n=6
arrangeCoins(n)
```

**Question 6**
Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]


```python
def Square_each(nums):
    # Square each number in the array and store the squares in a new list
    squares = [num ** 2 for num in nums]
    
    # Sort the squared numbers in non-decreasing order
    squares.sort()
    
    return squares
print("The time complexity is O(nlogn)")
print("Space COmplexity is O(n)")
nums = [-4,-1,0,3,10]
Square_each(nums)
```

**Question 7**
You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

Count and return *the number of maximum integers in the matrix after performing all the operations*

**Example 1:**

**Input:** m = 3, n = 3, ops = [[2,2],[3,3]]

**Output:** 4

**Explanation:** The maximum integer in M is 2, and there are four of it in M. So return 4.

```python
def maxCount(m, n, ops):
    if not ops:
        return m * n  # If ops is empty, return the total number of elements in the matrix

    # min_row = float('inf')  # Initialize min_row with infinity
    # min_col = float('inf')  # Initialize min_col with infinity

    for op in ops: # [2,2] means(0,0),(0,1),(1,0),(1,1) we need to increment this value
        m = min(m, op[0]) # Find the minimum value in the first column of ops
        n = min(n, op[1])  # Find the minimum value in the second column of ops

    return m * n  # Return the product of the minimum values found
m = 3
n = 3
ops = [[2, 2]]
maxCount(m,n,ops)
```

```python
print("The time complexity is O(ops)")
print("Space COmplexity is O(1)")
m = 3
n = 3
ops = [[2, 2],[3,3]]
maxCount(m,n,ops)
```

 **Question 8**

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

*Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

**Example 1:**

**Input:** nums = [2,5,1,3,4,7], n = 3

**Output:** [2,3,5,4,1,7]

**Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

</aside>

```python
def shuffle_2n(nums, n):
    result = []  # Initialize an empty list to store the shuffled array

    # Iterate from 0 to n (exclusive) to handle the x values
    for i in range(n):
        result.append(nums[i])  # Append the x value at index i
        result.append(nums[i + n])  # Append the corresponding y value at index i + n

    return result  # Return the shuffled array
```

```python
print("The time complexity is O(n)")
print("Space COmplexity is O(n)")
nums = [2,5,1,3,4,7]
n = 3
shuffle_2n(nums,n)
```
